```python
# you should not import any other packages
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from sklearn.linear_model import SGDRegressor
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
X= b * np.sin(phi)
Y= a * np.cos(phi)
X = X.reshape(50,1)
Y = Y.reshape(50,1)
print(X.shape,Y.shape)
```

```python
#Code starts here

alpha=[0.0001, 1, 100]
outliers = [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
plt.figure(figsize=(20,16))
plt.title('Regression line w.r.t Outliers')
plt.xlabel('Number of Outliers')
plt.ylabel('Alpha value')
for idx,alp in enumerate(alpha):
    X_new = X #saving the orgiginal set of points for each iterations
    Y_new = Y
    for j,i in enumerate(outliers):
        plt.subplot(3, 5, len(outliers)*idx+j+1)
        clf = SGDRegressor(alpha=alp, eta0=0.001, learning_rate='constant',random_state=0)
        X_new = np.append(X_new,i[0]).reshape(-1,1)#adding new outlier's X-cordinate value to the data set
        Y_new = np.append(Y_new,i[1]).reshape(-1,1)#adding new outlier's Y-cordinate value to the data set
        clf.fit(X_new,Y_new)#fitting the model with the updated X_train values
        y_pred = clf.predict(X_new)
        plt.scatter(X_new,Y_new)
        plt.plot(X_new,y_pred,color='red')#Plotting the regression Line
plt.show()
        
```

# Observations


1. First observation is that even the presence of a single outlier does impact the regression line.


2. Regression line changes so as to fit the single/few outlier that are present in the dataset.


3. One key observation is as the alpha value increases the regression line tends to change very slowly w.r.t to the outliers as compared to the model having a smaller alpha value. E.g with alpha value 100, upon adding just one outlier , the regression line does not change much inorder to fit the outlier as it did when alpha value was 0.0001 and 1.
